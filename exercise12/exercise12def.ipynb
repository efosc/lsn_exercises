## Exercise 12

The aim of Lecture 12 is to employ neural networks in a classification problem: recognize handwritten digits from 0 to 9.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import SGD, RMSprop, Adadelta, Adagrad
from tensorflow.keras.layers import Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.datasets import mnist
from tensorflow.keras.optimizers import SGD, Adam, RMSprop, Adagrad, Adadelta
import os
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from PIL import Image
```

### 12.1 

In this part the same DNN model was compiled using different optimizers and algorithm performances were compared.

```python
## cose generali
seed = 0
np.random.seed(seed)
tf.random.set_seed(seed)

img_rows, img_cols = 28, 28 #input image dimensions - number of pixels 
num_classes = 10 # 10 digits - output dimension (10 differet categories) cioè da 0 a 9

#the data, split between train and test sets
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()

print('X_train shape:', X_train.shape)
print('Y_train shape:', Y_train.shape)
print()
print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')
```

```python
#reshape data
X_train = X_train.reshape(X_train.shape[0], img_rows*img_cols)
X_test = X_test.reshape(X_test.shape[0], img_rows*img_cols)

#cast floats to single precesion
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

#rescale data in interval [0,1] - X are in [0,255]
X_train /= 255
X_test /= 255

#convert class vectors to binary class matrices, e.g. for use with categorical_crossentropy
print('an example of a data point with label', Y_train[20], 'before to_categorical ...')
Y_train = keras.utils.to_categorical(Y_train, num_classes)
Y_test = keras.utils.to_categorical(Y_test, num_classes)
print('... and with label', Y_train[20], 'after to_categorical')
print()
print('X_train shape:', X_train.shape)
print('Y_train shape:', Y_train.shape)
```

```python
## create DNN model

def create_DNN():
    # instantiate model
    model = Sequential()
    # add a dense all-to-all relu layer
    model.add(Dense(400,input_shape=(img_rows*img_cols,), activation='relu'))
    # add a dense all-to-all relu layer
    model.add(Dense(100, activation='relu'))
    # apply dropout with rate 0.5
    model.add(Dropout(0.5))
    # soft-max layer
    model.add(Dense(num_classes, activation='softmax'))
    
    return model
```

### A) $\rightarrow$ SGD optimizer

```python
## compile model (optimizer = SGD)
def compile_model_sgd():
    # create the model
    model=create_DNN()
    # compile the model
    model.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer=SGD(),
                  metrics=['accuracy'])
    return model
```

```python
## model with sgd numclasses=10
batch_size = 32
epochs = 15

# create the deep neural net
model_DNN_sgd = compile_model_sgd()

# train DNN and store training info in history
history_sgd = model_DNN_sgd.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          shuffle=True,
          validation_data=(X_test, Y_test))
```

```python
# evaluate model
score = model_DNN_sgd.evaluate(X_test, Y_test, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# summarize history for accuracy
plt.plot(history_sgd.history['accuracy'])
plt.plot(history_sgd.history['val_accuracy'])
plt.ylabel('model accuracy', size=12)
plt.xlabel('epoch',size=12)
plt.title("Model accuracy during training",size=12)
plt.legend(['train', 'test'], loc='best')
plt.grid(True)
plt.show()

# summarize history for loss
plt.plot(history_sgd.history['loss'])
plt.plot(history_sgd.history['val_loss'])
plt.ylabel('model loss')
plt.xlabel('epoch')
plt.title("Model loss function during training", size=12)
plt.legend(['train', 'test'], loc='best')
plt.grid(True)
plt.show()
```

```python
predictions = model_DNN_sgd.predict(X_test)

X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols,1)

plt.figure(figsize=(15, 15)) 
for i in range(10):    
    ax = plt.subplot(2, 10, i + 1)    
    plt.imshow(X_test[i, :, :, 0], cmap='gray')    
    plt.title("Digit: {}\nPredicted:    {}".format(np.argmax(Y_test[i]), np.argmax(predictions[i])))    
    plt.axis('off') 
plt.show()
```

### 12.2

In this section I created a CNN model with:
- two pairs of convolutional + pooling layers (first convolutional layer with 10 neurons, second with 14 neurons)
- a dense layer with 35 neurons and *relu* activation function 
- a final dense layer with with *relu* activation function

The model reaches $\approx 98$% accuracy with a 20-epochs training. 

```python
#reshape data, depending on Keras backend
if keras.backend.image_data_format() == 'channels_first':
    X_train = X_train.reshape(X_train.shape[0], 1, img_rows, img_cols)
    X_test = X_test.reshape(X_test.shape[0], 1, img_rows, img_cols)
    input_shape = (1, img_rows, img_cols)
else:
    X_train = X_train.reshape(X_train.shape[0], img_rows, img_cols, 1)
    X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols, 1)
    input_shape = (img_rows, img_cols, 1)
    
print('X_train shape:', X_train.shape)
print('Y_train shape:', Y_train.shape)
print('X_test shape:', X_test.shape)
print('Y_test shape:', Y_test.shape)
print()
print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')
```

```python
def create_CNN():
    # instantiate model
    model = Sequential()
    # add first convolutional layer with 10 filters (dimensionality of output space)
    model.add(Conv2D(10, kernel_size=(5, 5),
                     activation='relu',
                     input_shape=input_shape))
    model.add(MaxPooling2D(pool_size=(2, 2), strides=2))
    model.add(Conv2D(14, kernel_size=(5,5),
                     strides=(1,1),
                     activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2), strides=2))
    model.add(Flatten())
    model.add(Dense(35, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(num_classes, activation='softmax'))
    
    # compile the model
    model.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer='SGD',
                  metrics=['accuracy'])
    return model
```

```python
batch_size = 50
epochs = 20

# create the deep conv net
model_CNN = create_CNN()
model_CNN.summary()

# train CNN
history_CNN = model_CNN.fit(X_train, Y_train,
              batch_size=batch_size,
              epochs=epochs,
              verbose=1,
              shuffle=True,
              validation_data=(X_test, Y_test))
```

```python
# evaluate model
score = model_CNN.evaluate(X_test, Y_test, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# summarize history for accuracy
plt.plot(history_CNN.history['accuracy'])
plt.plot(history_CNN.history['val_accuracy'])
plt.ylabel('model accuracy', size=12)
plt.xlabel('epoch',size=12)
plt.title("Model accuracy during training",size=12)
plt.legend(['train', 'test'], loc='best')
plt.grid(True)
plt.show()

# summarize history for loss
plt.plot(history_CNN.history['loss'])
plt.plot(history_CNN.history['val_loss'])
plt.ylabel('model loss')
plt.xlabel('epoch')
plt.title("Model loss function during training", size=12)
plt.legend(['train', 'test'], loc='best')
plt.grid(True)
plt.show()
```

```python
predictions = model_CNN.predict(X_test)

X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols,1)

plt.figure(figsize=(15, 15)) 
for i in range(10):    
    ax = plt.subplot(2, 10, i + 1)    
    plt.imshow(X_test[i, :, :, 0], cmap='gray')    
    plt.title("Digit: {}\nPredicted:    {}".format(np.argmax(Y_test[i]), np.argmax(predictions[i])))    
    plt.axis('off') 
plt.show()
```

### 12.3

Finally the CNN model from the previuos section was applied to new handwritten digits. The algorithm only fails in recognizing the last digit (9).

```python
## load digits
datatot = []
for k in range(10):
    digit_filename = str(k) + ".png"
    digit_in = Image.open(digit_filename).convert('L')

    ydim, xdim = digit_in.size
    #print("Image size: "+str(xdim)+"x"+str(ydim))
    pix=digit_in.load();
    data = np.zeros((xdim, ydim))
    for j in range(ydim):
        for i in range(xdim):
            data[i,j]=pix[j,i]

    data /= 255
    
    datatot.append(data)
```

```python
predictions = []
for data in datatot:
    data = data.reshape(1,28,28,1)
    prediction = model_CNN.predict(data)
    predictions.append(prediction)
    
plt.figure(figsize=(15, 15)) 
for i in range(10):    
    ax = plt.subplot(2, 10, i + 1)    
    plt.imshow(datatot[i], cmap='gray')    
    plt.title("Digit: " + str(i) + "\nPredicted: " + str(np.argmax(predictions[i])))    
    plt.axis('off') 
plt.show()
```
