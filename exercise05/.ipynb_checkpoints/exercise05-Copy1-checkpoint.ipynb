## Exercise 5

### Introduction

In this exercise the Metropolis algorithm was applied to sample the probability distribution of the orbitals 1s and 2p of the Hydrogen atom. Lengths are expressed in Bohr radius ($a_0 = 1$) units, so that $r$ results adimensional. With $a_0 = 1$ the expressions of $\Psi$ and $|Psi|^2$ become:
$$
\Psi_{1,0,0}(r,\theta,\phi)= \frac{1}{\sqrt{\pi}} e^{-r} \qquad \rightarrow  \qquad 
|\Psi_{1,0,0}(r,\theta,\phi)|^2 = \frac{1}{\pi}e^{-2r}
$$

$$
\Psi_{2,1,0}(r,\theta,\phi)=
\frac{1}{8}\sqrt{\frac{2}{\pi}} r e^{-r/2} \cos(\theta) \qquad \rightarrow  \qquad
|\Psi_{2,1,0}(r,\theta,\phi)|^2 = \frac{1}{32 \pi}r^2 e^{-r} \cos^2(\theta)
$$

In the first part of the exercise the algorithm parameters were tuned to optimize the code. On the other hand, the second part of the notebook reports the estimations of $\left\langle r \right\rangle_{\Psi_{1,0,0}}$ $\left\langle r \right\rangle_{\Psi_{2,1,0}}$ (compared to their expectation values $
\left\langle r \right\rangle_{\Psi_{1,0,0}} = \frac{3}{2}a_0$
 and 
$\left\langle r \right\rangle_{\Psi_{2,1,0}} = 5 a_0
$
, as well as a 3D representation of the orbitals.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

### Part 1: Optimization - parameter tuning

In this first part of the excercise, I ran some preliminary simulations in order to correctly tune the Metropolis and data-blocking parameters. 
More specifically, these three parameters were taken into account:
1. $\delta$ (i.e. the distance between to subsequent different positions, also referred to as *step length* from now on),
2. the simulation starting point,
3. the block legth (number of algorithm steps in each block) during data blocking.

#### Step length optimization

$\rightarrow$ *ottimizzazione_nuovo.cpp*

At first, I looked for the best value of $\delta$ in order to obtain the 50% of acceptance.
To do so, I ran some simulations varying $\delta$ in steps of 0.01, from 0.01 to a maximum,  which I choose as equal to the distance between the simulation starting position and the maximum probability position. I repeated these simulations with different starting points (near or away from the maximum of the probability distribution).

The following graphs report examples of the accepted steps trend for s/p orbitals and uniform/gauss transition probability (only the $\delta \in [0.01, 5.00]$ part is shown). 

```python
M=500
## questo sempre lo stesso
step = np.loadtxt( './Optimization/accepted_s_gauss.dat',max_rows=M, usecols = 1 )

acc_s_unif = np.loadtxt('./Optimization/accepted_s_uniform_prova2.dat',max_rows=M, usecols=4)
acc_s_gauss = np.loadtxt( './Optimization/accepted_s_gaussprova4.dat',max_rows=M, usecols=4 )
acc_p_unif = np.loadtxt( './Optimization/accepted_p_uniform_prova2.dat',max_rows=M, usecols=4)
acc_p_gauss = np.loadtxt( './Optimization/accepted_p_gauss_prova3.dat',max_rows=M, usecols=4)

plt.figure(figsize=(15,15))
plt.suptitle('Number of accepted positions as a function of the step dimension', size=17)

plt.subplot(221)
plt.plot(step,acc_s_unif)
plt.title("s, uniform, $\delta =$ 1.24", size=17)
plt.ylabel('# accepted positions over 10000',size=14)
plt.xlabel('step dimension',size=14)
plt.hlines(5000, xmin=0., xmax=max(step), color="red", label="50% acceptance")
plt.legend(loc="upper right", fontsize=14)
plt.grid(True)

plt.subplot(222)
plt.plot(step,acc_s_gauss)
plt.title("s, gauss, best $\delta =$ 0.79", size=17)
plt.ylabel('# accepted positions over 10000',size=14)
plt.xlabel('step dimension',size=14)
plt.hlines(5000, xmin=0., xmax=max(step), color="red", label="50% acceptance")
plt.legend(loc="upper right", fontsize=14)
plt.grid(True)

plt.subplot(223)
plt.plot(step,acc_p_unif)
plt.title("p, uniform, best $\delta =$ 2.94", size=17)
plt.ylabel('# accepted positions over 10000',size=14)
plt.xlabel('step dimension',size=14)
plt.ylabel('# accepted positions over 10000',size=14)
plt.hlines(5000, xmin=0., xmax=max(step), color="red", label="50% acceptance")
plt.legend(loc="upper right", fontsize=14)
plt.grid(True)

plt.subplot(224)
plt.plot(step,acc_p_gauss)
plt.title("p, gauss, best $\delta =$ 1.88", size=17)
plt.ylabel('# accepted positions over 10000',size=14)
plt.xlabel('step dimension',size=14)
plt.ylabel('# accepted positions over 10000',size=14)
plt.hlines(5000, xmin=0., xmax=max(step), color="red", label="50% acceptance")
plt.legend(loc="upper right", fontsize=14)
plt.grid(True)
```

The number of accepted steps seems not to depend significantly on the starting position (the best value of $\delta$ varies only at the second decimal digit). I choose not to improve the optimization any further (considering the third decimal digit), since even the second decimal digit depends on the starting simulation point. I obtained the following $\delta$ values:

1. s orbital, uniform transition probability $\rightarrow$ $\delta =$ 1.22,  accepted steps = 4996,  error $= \frac{|4996-5000|}{5000} =$ 0.0008 
2. s orbital, gauss transition probability $\rightarrow$ $\delta =$ 0.79,  accepted steps = 5010, error $= \frac{|5010-5000|}{5000} =$ 0.002
3. p orbital, uniform transition probability $\rightarrow$ $\delta =$ 2.94,  accepted steps = 5017, error $= \frac{|5017-5000|}{5000} =$0.0034
4. p orbital, gauss transition probability $\rightarrow$ $\delta =$ 1.88,  accepted steps = 4999,  error $= \frac{|4999-5000|}{5000} =$ 0.0002

#### Starting point

$\rightarrow$ *instant.cpp*

For the s orbital, the distance from the origin approaches the expected value $
\left\langle r \right\rangle_{\Psi_{1,0,0}} = \frac{3}{2}a_0 $ even if the simulation is started  away from the origin. However this is not true for the p orbital: here the *r* value keeps fluctuating between 0 and 15$a_0$, even after running the simulation for a large number of steps. This could be due to the fact that the $\rho_s = |\Psi_{1,0,0}(r,\theta,\phi)|^2$ probability distribution has just one maximum in (0,0,0), while $\rho_p = |\Psi_{2,1,0}(r,\theta,\phi)|^2$ has two maxima in (0,0,5) and (0,0,-5). The following graphs represent the $r$ instant values as a function of the number of steps, when starting away from the origin (and from the expected $\langle r \rangle$ value). They were useful for evaluating the equilibration time. 

```python
pos_u = np.loadtxt( './Optimization/positions_s_uniform.dat', usecols=4 )
pos_g = np.loadtxt( './Optimization/positions_s_gauss.dat', usecols=4 )
step = np.loadtxt( './Optimization/positions_s_uniform.dat', usecols = 0 ) 

plt.figure(figsize=(15,6))
plt.plot(step,pos_u, label = "Uniform transition probability")
plt.plot (step, pos_g, color="midnightblue", label = "Gauss transition probability")
plt.hlines( 1.5, xmin = 0, xmax=max(step), color='red', label="Expected value")
plt.xlabel('# steps',size=14)
plt.ylabel('Distance from the origin [$a_0$ units]',size=14)
plt.title("Radius as a function of the number of steps, s orbital", size=17)
plt.legend(loc="upper right", fontsize=14)
plt.grid(True)
```

#### Block length

Since the total number of steps is fixed to  $M = 10^6$, I set the block length to $L=5000$ and the number of blocks to $N=200$. Indeed, averages within each block are uncorreated if the blocks are large enough. However, I noteced that increasing the number of steps per block (10000) the final error bars did not improve, so I considered 5000 as a reasonable block length.

### Part 2: Results

$\rightarrow$ *medie_nuovo.cpp*

The following graphs show the estimations of $\left\langle r \right\rangle_{\Psi_{1,0,0}}$ and $\left\langle r \right\rangle_{\Psi_{2,1,0}}$</span> and their uncertainties as a function of the number of blocks.

```python
## questa è una altra prova di s, gauss con step tot = 10^6, step nel blocco = 5000

pos = np.loadtxt( './Blocking/averagess_gauss_volta4.dat', usecols=3 )
step = np.loadtxt( './Blocking/averagess_gauss_volta4.dat', usecols = 0 )
err = np.loadtxt( './Blocking/averagess_gauss_volta4.dat', usecols = 4 )
pos2 = np.loadtxt( './Blocking/averagess_uniform_volta1.dat', usecols=3 )
err2 = np.loadtxt( './Blocking/averagess_uniform_volta1.dat', usecols = 4 )
step = step*1000

plt.figure(figsize=(15,6))
plt.errorbar(step, pos, err, label = "gauss")
plt.errorbar(step, pos2, err2, label = "uniform")
plt.hlines( 1.5, xmin = 0, xmax=max(step), color='red', label="Expected value")
plt.xlabel('# steps',size=14)
plt.ylabel('distance from the origin',size=14)
plt.title("Radius as a function of the number of blocks", size=17)
plt.suptitle("s orbital", size=17, weight="bold" )
plt.legend(loc="upper right", fontsize=14)
plt.grid(True)
```
